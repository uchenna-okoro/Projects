## Clustering Univariate, Bivariate & Multivariate

```python
clustering=KMeans(n_clusters=6)
```

```python
clustering.fit(mall_dataDF[['Annual Income (k$)']])
```

```python
clustering.labels_
```

```python
mall_dataDF['Income Cluster']=clustering.labels_
```

```python
mall_dataDF['Income Cluster'].head()
```

```python
mall_dataDF['Income Cluster'].value_counts()
```

```python
clustering.inertia_
```

```python
inertia_scores=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(mall_dataDF[['Annual Income (k$)']])
    inertia_scores.append(kmeans.inertia_)
    
    
```

```python
inertia_scores
```

```python
plt.plot(range(1,11),inertia_scores)
```

## Bivariate Clustering

```python
clustering2=KMeans()
clustering2.fit(mall_dataDF[['Annual Income (k$)',
       'Spending Score (1-100)']])
mall_dataDF['Spending and Income Cluster']=clustering2.labels_
mall_dataDF.head()

```

```python
inertia_scores=[]
for i in range(1,11):
    kmeans2=KMeans(n_clusters=i)
    kmeans2.fit(mall_dataDF[['Spending and Income Cluster']])
    inertia_scores.append(kmeans2.inertia_)
plt.plot(range(1,11),inertia_scores)
```

```python
#looking for centroids
centers=pd.DataFrame(clustering2.cluster_centers_)
centers.columns=['x','y']
centers.head(10)
```

```python
# using scatter plot
plt.figure(figsize=(10,8))
plt.scatter(x=centers['x'],y=centers['y'],s=100,c='black',marker='*')
sns.scatterplot(data=mall_dataDF,x='Annual Income (k$)',y='Spending Score (1-100)',hue='Spending and Income Cluster',palette='tab10')
```

```python
pd.crosstab(mall_dataDF['Spending and Income Cluster'],mall_dataDF['Genre'],normalize='index')
```

```python
mall_dataDF.groupby('Spending and Income Cluster')[ 'Age', 'Annual Income (k$)',
       'Spending Score (1-100)'].mean()
```

### Multivariate Clustering

```python
#import standard scaler and do preprocessing
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
mall_dataDF.head()
```

```python
#getting new dataframe
mall_dataDF=pd.get_dummies(mall_dataDF)
mall_dataDF.head()
```

```python
#identify all categorical variables
categorical_columns = mall_dataDF.select_dtypes(['object']).columns

#convert all categorical variables to numeric
mall_dataDF[categorical_columns] = mall_dataDF[categorical_columns].apply(lambda x: pd.factorize(x)[0])
mall_dataDF.head()
```

```python
mall_dataDF=pd.get_dummies(mall_dataDF,drop_first=True)
mall_dataDF.head()
```

```python

''' drop multiple column based on name'''
mall_dataDF.drop(['CustomerID', 'Income Cluster','Genre_Female'], axis = 1).head()
```

```python
mall_dataDF=scale.fit_transform(mall_dataDF)
```

```python
mall_dataDF=pd.DataFrame(scale.fit_transform(mall_dataDF))
```

```python
mall_dataDF.head()
```

```python
inertia_scores3=[]
for i in range(1,11):
    kmeans3=KMeans(n_clusters=i)
    kmeans3.fit(mall_dataDF)
    inertia_scores3.append(kmeans3.inertia_)
plt.plot(range(1,11),inertia_scores3)
```
